```python
# Importing the Necessary Libraries
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
img = cv2.imread(r"C:\Users\Ebi\Desktop\Task 4 LGM\Covert_Image_To_Pencil_Sketch\img4.jpg")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

By using the cv2.cvtColor() function with the cv2.COLOR_BGR2RGB conversion code, I am converting the image img from BGR to RGB color space.

```python
display(Image.fromarray(img))
```

# Converting into Gray Scale

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

 This converts an image img from the BGR color space to grayscale using OpenCV's cv2.cvtColor() function with the cv2.COLOR_BGR2GRAY conversion code. Then, it uses the Image.fromarray() function from the PIL library to display the resulting grayscale image.

# Blurring the image

```python
# Invert the grayscale image
inv_gray_image = 255 - gray_image
# Display the inverted image
display(Image.fromarray(inv_gray_image))
```

It calculates the inverted grayscale image by subtracting each pixel value from 255 (assuming 8-bit pixel values). Then, it uses the Image.fromarray() function to display the resulting inverted grayscale image.

```python
# Apply Gaussian blur to the inverted grayscale image
blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
```

```python
display(Image.fromarray(inv_gray_image))
```

```python
inverted_blur_image = 255 - blur_image
```

## Pencil Sketch Image is as follows

```python
pencil_sketch = cv2.divide(gray_image, inverted_blur_image, scale = 256.0)
```

```python
cv2.imshow("Sketch Image", pencil_sketch)
display(Image.fromarray(pencil_sketch))
```

```python
# Display the original image
display(Image.fromarray(img))

# Display the pencil sketch
display(Image.fromarray(pencil_sketch))
```
